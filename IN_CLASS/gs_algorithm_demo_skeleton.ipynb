```python
# adjaceny connectivity list for graph
A = [[2,3],
    [4],
    [],
    [3]]
A = [[x-1 for x in row] for row in A];

# adjaceny weight list for graph
W = [[1,4],
     [1],
     [],
     [1]];
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# visualize graph with weights
G = nx.Graph()
for i in range(len(A)):
    for j in range(len(A[i])):
        G.add_edge(i, A[i][j], weight=W[i][j])

pos = nx.spring_layout(G)  # generate a layout and store it in pos
nx.draw(G, pos, with_labels=True)  # use pos as the position argument
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)  # use pos as the position argument

plt.show()
# By generating the layout once and reusing it, you ensure that the positions of the nodes and edges are consistent between the drawing of the graph and the drawing of the edge labels.
```
