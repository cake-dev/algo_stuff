```python
# Retunrs an NxN array of preferences for men
#  and woment that has been randomly generated
#
#  Input:
#     N - number of men/women in preference list

def generate_random_preferences(N):

  # CREATE PREFERENCE LISTS
  pref_m = np.zeros((N,N),dtype=np.int8)
  pref_w = np.zeros((N,N),dtype=np.int8)
  # Intialize some preference lists
  for i_person in np.arange(N):
    pref_m[i_person] = np.random.permutation(np.arange(N))
    pref_w[i_person] = np.random.permutation(np.arange(N))

  return pref_m,pref_w
```

```python
# Function to "invert" preference list.  We want an array where
#  we index into array by person's ID and it returns their rank
#            i.e.,
#           pref_w[rank_of_m] = m
#      pref_w_inv[m] = rank_of_m
#
#
#  Input:
#     pref_array - an NxN np array of preferences
#
#  Output:
#     pref_array_inv - an "inverted" NxN np array
#

def invert_pref_array(pref_array):

  # Get size of the problem
  N = pref_array.shape[0];

  # Preprocessing - invert the female preference list, O(n) time
  pref_array_inv = np.zeros((N,N))
  # Invert female preference list
  for i in np.arange(N):
      for j in np.arange(N):
        pref_array_inv[i,pref_array[i,j]]=j

  return pref_array_inv
```

```python

# IMPLEMENTATION OF GS ALGORITHM
#  Assume we have men with indices 0...N-1
#  Assume we have women with indices 0'...N-1'
#
# INPUT:
#     pref_m - np array of men indices, in descending preference
#     pref_w - np array of indices, in descending preference
#
# OUTPUT
#     husband - np array of husband for each woman
#     wife - np array of wife for each husband

def gs_algo(pref_m,pref_w):

  #Get size of the problem
  N = pref_w.shape[0];

  # Create arrays for husband and wife tracking
  wife = -1*np.ones(N,dtype=np.int8)
  husband = -1*np.ones(N,dtype=np.int8)

  # Keep track of where each man is on his list
  m_pointer = np.zeros(N,dtype=np.int8)

  # Invert female preferences
  pref_w_inv = invert_pref_array(pref_w);

  # Create a stack of free men
  free_men = np.arange(N,dtype=np.int8)
  # free_men = np.random.permutation(free_men);  # Randomly shuffle
  free_men = np.ndarray.tolist(free_men)  # Convert to list

  # While there are free men, do the following
  iter = 0;
  while len(free_men) != 0:
    iter = iter+1
    print('Iteration  ' + str(iter))

    m = free_men.pop(0);  #Pop top man off the stack
    w_selected = pref_m[m,m_pointer[m]];  #Woman to propose to
    m_pointer[m]+=1 # increment the pointer

    # ipdb.set_trace()

    # Check if preferred woman is paired
    if husband[w_selected]==-1:

        #If unpaired, make m her husband
        husband[w_selected] = m;
        wife[m] = w_selected;

        #Display what happened
        print('Wife ' + str(w_selected) + ' occupied by man ' + str(m))

    else:  #There is a man there

        m_current = husband[w_selected]

        # ipdb.set_trace()

        #Compare men ranks
        if pref_w_inv[w_selected,m] < pref_w_inv[w_selected, m_current ]:
          # If m wins, replace current husband
          husband[w_selected] = m;
          wife[m] = w_selected;

          # Put displaced man onto stack
          free_men.insert(0,m_current)

          # Dispaly what happened
          print('Wife ' + str(w_selected) + ' current man ' + str(m_current) + ' displaced by man ' + str(m))

        else:
          #If m loses, put him back on the stack
          free_men.insert(0,m)

          #Display what happened
          print('Wife ' + str(w_selected) + ' current man ' + str(m_current) + ' NOT displaced by man ' + str(m))

  return husband,wife

```
